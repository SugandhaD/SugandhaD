## Q 5. Find the net revenue generated by the E-Commerce company over all orders placed. (6 marks)

### Ans 5 
Option-1 solving the question by using for loop and if,elif and else condition .

```python
l=[]
for i in (ECom_Data['discounted_price']):
                            if (i > 600):
                                l.append(0.25*i)
                            elif (i >350)  and (i <=600):
                                l.append(0.15*i)
                            elif (i >100) and (i <=350):
                                l.append(0.10*i)
                            else :
                                l.append(0.05*i)

ECom_Data['Ecom_Revenue']=l
ECom_Data.head()
```

```python
#net revenue generated by the E-Commerce company over all orders.
ECom_Data['Ecom_Revenue'].sum()
```

### Ans 5 
Option-2 solving the question by using lambda function and - and ,or condition.

```python
df_Ecom=ECom_Data.copy()
df_Ecom['Ecom_Revenue'] = df_Ecom['discounted_price'].apply(
    lambda x: (x > 600 and x * .25) or (x > 350 and x *.15) or (x > 100 and x *.10) or x*.05)
df_Ecom.head()
```
